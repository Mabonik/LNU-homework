## Задача I
Дано три цілих числа. Найдіть найбільше з них (програма повинна вивести рівно одне ціле число).

```python
a = int(input("a: "))
b = int(input("b: "))
c = int(input("c: "))
if c>a and c>b:
    print(c)
elif a>b and a<c:
    print(a)
else:
    print(b)
```

## Задача J
Дано три натуральних числа a, b, c. Визначте, чи існує трикутник з такими сторонами. Якщо трикутник існує, виведіть "ТАК", а якщо не є трикутником "НІ".

```python
a = int(input())
b = int(input())
c = int(input())
a, b, c = sorted([a, b, c])
if a + b <= c:
    print('НІ')
else:
    print('ТАК')
```
